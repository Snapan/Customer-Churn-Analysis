# Converting 1 and 0 values of Senior Citizen to yes/no for easy comprehension

```python
def conv(value):
  if value == 1:
    return "Yes"
  else:
    return "No"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
df.head(50)
```

```python
plt.figure(figsize=(5,4))
ax=sns.countplot(x='Churn', data=df)
ax.set_title('Churn Count')
ax.bar_label(ax.containers[0])
ax.set_xlabel('Churn')
ax.set_ylabel('Count')
plt.show()
```

```python
crosstab = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plot stacked bar chart
ax = crosstab.plot(kind='bar', stacked=True, figsize=(5, 4), color=['#1f77b4', '#ff7f0e'])

# Add percentages on the bars
for p in ax.patches:
    width = p.get_width()    # bar width
    height = p.get_height()  # bar height
    x, y = p.get_xy()        # bar bottom-left corner
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

# Add title and labels
plt.title("Churn by SeniorCitizen (Stacked Bar with %)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn")

# Show the plot
plt.show()
```

```python

# Assuming you have a DataFrame called df with the specified columns
# Sample DataFrame creation for example
# Columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the matplotlib figure and axes (3 rows, 3 columns for a 3x3 grid)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# Flatten the axes for easier access
axes = axes.flatten()

# Create a countplot for each column
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i],hue ="Churn")
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

# The plots suggest that customers without added services like online security, tech support, and device protection have higher churn rates compared to those who use these services. Additionally, customers using fiber optic internet appear more likely to churn than those with DSL. Generally, customers without internet service or minimal added services tend to have lower churn rates.
